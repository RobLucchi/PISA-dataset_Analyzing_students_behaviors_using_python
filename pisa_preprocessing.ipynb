```python
selected_columns = ['AGE','NC','ST04Q01',
           'ST11Q01','ST11Q02','ST11Q03','ST11Q04',
           'ST19Q01','ST15Q01',
           'ST44Q03','ST44Q04','ST44Q07','ST44Q08',
           'ST87Q01','ST87Q04','ST87Q06','ST87Q09',
           'TEACHSUP','MATWKETH']

df = pisa[selected_columns]
df.head()
```
